```python
# импортируем библиотеку
import random
# приветствие
print("Привет, чтобы начать игру, введите 'start'; чтобы закончить игру, введите 'finish'; чтобы посмотреть счёт, введите 'score'")
# счетчик очков(словарь)
scr = {'PC': 0, 'You': 0}
# создаем цикл для беспрерывной работы алгоритма
a = "start"
while a != "finish":
  # а - ввод пользователя
  a = input()
  if a == "start":
    print("Введите 'sam', чтобы компьютер делал одинаковые ходы; введите 'rand', чтобы компьютер делал рандомные ходы; введите 'hard', чтобы компьюер играл в полную силу")
    # b - ввод пользователя(выбираем стиль игры компьютера)
    b = input()
    # режим игры sam
    if b == "sam":
      print("Для начала, укажите количество раундов")
      print("Чтобы сделать ход, во второй строке введите 'камень', 'ножницы' или 'бумага'")
      # программа рандомно выбирает фигуру для игры( 1 - камень; 2 - ножницы; 3 - бумага)
      # pc - ход программы
      pc = random.randint(1, 3)
      # кол-во раундов
      n = int(input())
      for i in range(n):
         # проверка с камнем
        if pc == 1:
          # pl - ход игрока
          pl = input()
          if pl == "ножницы":
            print("PC:", "камень")
            print("You:", pl)
            print("Вы проиграли(")
            # зачисление очков победителю( и в остальных подобных блоках)
            scr["PC"] += 1
          elif pl == "бумага":
            print("PC:", "камень")
            print("You:", pl)
            print("Вы выиграли)")
            scr["You"] += 1
          elif pl == "камень":
            print("PC:", "камень")
            print("You:", pl)
            print("Ничья")
          # защита от поломки кода
          else:
            print("Введите одну из предложенных команд")
        # проверка с ножницами
        elif pc == 2:
          pl = input()
          if pl == "бумага":
            print("PC:", "ножницы")
            print("You:", pl)
            print("Вы проиграли(")
            scr["PC"] += 1
          elif pl == "камень":
            print("PC:", "ножницы")
            print("You:", pl)
            print("Вы выиграли)")
            scr["You"] += 1
          elif pl == "ножницы":
            print("PC:", "ножницы")
            print("You:", pl)
            print("Ничья")
          else:
            print("Введите одну из предложенных команд")
        # проверка с бумагой
        elif pc == 3:
          pl = input()
          if pl == "камень":
            print("PC:", "бумага")
            print("You:", pl)
            print("Вы проиграли(")
            scr["PC"] += 1
          elif pl == "ножницы":
            print("PC:", "бумага")
            print("You:", pl)
            print("Вы выиграли)")
            scr["You"] += 1
          elif pl == "бумага":
            print("PC:", "бумага")
            print("You:", "бумага")
            print("Ничья")
          else:
            print("Введите одну из предложенных команд")
    # режим игры rand
    elif b == "rand":
      print("Для начала, укажите количество раундов")
      print("Чтобы сделать ход, во второй строке введите 'камень', 'ножницы' или 'бумага'")
      n = int(input())
      for i in range(n):
        # здесь программе необходимо выбирать фигуру в теле цикла( каждый раунд фигура рандомно обновляется)
        pc = random.randint(1, 3)
        # далее точно так же как и в sam
        if pc == 1:
          pl = input()
          if pl == "ножницы":
            print("PC:", "камень")
            print("You:", pl)
            print("Вы проиграли(")
            scr["PC"] += 1
          elif pl == "бумага":
            print("PC:", "камень")
            print("You:", pl)
            print("Вы выиграли)")
            scr["You"] += 1
          elif pl == "камень":
            print("PC:", "камень")
            print("You:", pl)
            print("Ничья")
          else:
            print("Введите одну из предложенных команд")
        elif pc == 2:
          pl = input()
          if pl == "бумага":
            print("PC:", "ножницы")
            print("You:", pl)
            print("Вы проиграли(")
            scr["PC"] += 1
          elif pl == "камень":
            print("PC:", "ножницы")
            print("You:", pl)
            print("Вы выиграли)")
            scr["You"] += 1
          elif pl == "ножницы":
            print("PC:", "ножницы")
            print("You:", pl)
            print("Ничья")
          else:
            print("Введите одну из предложенных команд")
        elif pc == 3:
          pl = input()
          if pl == "камень":
            print("PC:", "бумага")
            print("You:", pl)
            print("Вы проиграли(")
            scr["PC"] += 1
          elif pl == "ножницы":
            print("PC:", "бумага")
            print("You:", pl)
            print("Вы выиграли)")
            scr["You"] += 1
          elif pl == "бумага":
            print("PC:", "бумага")
            print("You:", "бумага")
            print("Ничья")
          else:
            print("Введите одну из предложенных команд")
    # режим игры hard
    elif b == "hard":
      print("Для начала, укажите количество раундов")
      print("Чтобы сделать ход, во второй строке введите 'камень', 'ножницы' или 'бумага'")
      n = int(input())
      # создаем счетчик ходов игрока для статистики
      plmovco = {"камень": 0, "ножницы": 0, "бумага": 0}
      for i in range(n):
        pl = input()
        # защита от поломки кода
        if pl not in plmovco:
          print("Введите одну из предложенных команд")
          continue
        # считаем ход игрока
        plmovco[pl] += 1
        # предсказание хода компьютера на основе предыдущих ходов игрока
        if not any(plmovco.values()):
          pc = random.randint(1, 3)
        else:
          # выбираем максимально часто используемую игроком фигуру и выбрасываем её противовес(фигуру, которая бьет фигуру игрока)
          mostcommov = max(plmovco, key=plmovco.get)
          pc = {"камень": 3, "ножницы": 1, "бумага": 2}[mostcommov]
        pc_choice = {1: "камень", 2: "ножницы", 3: "бумага"}[pc]
        print(f"PC: {pc_choice}")
        print(f"You: {pl}")
        # очередная проверка победителя
        if (pl == "камень" and pc_choice == "ножницы") or \
                (pl == "ножницы" and pc_choice == "бумага") or \
                (pl == "бумага" and pc_choice == "камень"):
          print("Вы выиграли)")
          scr["You"] += 1
        elif pl == pc_choice:
          print("Ничья")
        else:
          print("Вы проиграли(")
          scr["PC"] += 1
    else:
      print("Введите одну из предложенных команд")
  # окончание программы
  elif a == "finish":
    print("Спасибо за игру;)")
    exit()
  # счет
  elif a == "score":
    print("Счёт:")
    print(scr)
  else:
    print("Введите одну из предложенных команд")
```
